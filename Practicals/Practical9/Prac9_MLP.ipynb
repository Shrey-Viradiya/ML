```python
from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.neural_network import MLPClassifier, MLPRegressor
```

```python
def Report(clf, prediction, y_true, class_names):
    print("Prediction Report: ", clf.__name__)
    print("===================")

    print("Accuracy: ", metrics.accuracy_score(y_true, prediction))
    print(metrics.classification_report(y_true, prediction, target_names = class_names))
    print(metrics.confusion_matrix(y_true, prediction))
```

## Multilayer Layer Classifier

### Iris Dataset

```python
iris = datasets.load_iris()
class_names, feature_names = iris['target_names'], iris['feature_names']
X = iris['data']
y = iris['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=259)
```

```python
MLPClassifier?
```

```python
param_grid = {
    'hidden_layer_sizes' : [(10, 5),(20, 10),(20, 5),(10, 10)],
    'solver' : ['sgd', 'adam', 'lbfgs'],
    'learning_rate' : ['constant', 'invscaling', 'adaptive']
}

clf = MLPClassifier()
clf_grid = GridSearchCV(clf, param_grid, cv = 5)
clf_grid.fit(X_train, y_train)
clf_grid.__name__ = 'IRIS_MLP_Classifier'
print(f"Best Parameters: {clf_grid.best_params_}")
```

```python
prediction = clf_grid.predict(X_test)
```

```python
Report(clf_grid, prediction, y_test, class_names)
```

### Breast Cancer Dataset

```python
breast_cancer = datasets.load_breast_cancer()
class_names, feature_names = breast_cancer['target_names'], breast_cancer['feature_names']
X = breast_cancer['data']
y = breast_cancer['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=259)
```

```python
param_grid = {
    'hidden_layer_sizes' : [(35, 5),(35, 10),(35, 20),(35, 20, 10)],
    'solver' : ['sgd', 'adam', 'lbfgs'],
    'learning_rate' : ['constant', 'invscaling', 'adaptive']
}

clf = MLPClassifier()
clf_grid = GridSearchCV(clf, param_grid, cv = 5)
clf_grid.fit(X_train, y_train)
clf_grid.__name__ = 'Breast_Cancer_Dataset'
print(f"Best Parameters: {clf_grid.best_params_}")
```

```python
prediction = clf_grid.predict(X_test)
```

```python
Report(clf_grid, prediction, y_test, class_names)
```

## Multilayer Layer Regressor

### Boston Dataset

```python
boston = datasets.load_boston()
feature_names = boston['feature_names']
X = boston['data']
y = boston['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=259)
```

```python
MLPRegressor?
```

```python
param_grid = {
    'hidden_layer_sizes' : [(35, 5),(35, 10),(35, 20),(35, 20, 10)],
    'solver' : ['sgd', 'adam', 'lbfgs'],
    'learning_rate' : ['constant', 'invscaling', 'adaptive']
}

reg = MLPRegressor()
reg_grid = GridSearchCV(reg, param_grid, cv = 5)
reg_grid.fit(X_train, y_train)
reg_grid.__name__ = 'Boston_Regressor'
print(f"Best Parameters: {reg_grid.best_params_}")
```

```python
prediction = reg_grid.predict(X_test)
```

```python
print("MAE: ", metrics.mean_absolute_error(y_test, prediction))
print("MSE: ", metrics.mean_squared_error(y_test, prediction))
```
